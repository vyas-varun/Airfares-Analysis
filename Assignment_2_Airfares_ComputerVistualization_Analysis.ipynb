**Dataframe creation with statitical infos on each starting city.**

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}

global_stats = Airfares['FARE'].groupby(Airfares['S_CITY']).apply(get_stats).unstack()
global_stats = global_stats.sort_values('count')
global_stats
```

```python
global_stats.plot(legend=True)
plt.title("Statistical analysis of starting city")
```

```python
global_stats.plot.barh(figsize=(11,12), stacked=False)
plt.show()
```

Insights: Vacation analysis with the coupon (no stop flights are prefered in off seasons where as flights with multiple stops are considered in vacation period). Moreover, fare for the flight is also less. 

```python
VacationAirfares = VacationAirfares[ (VacationAirfares['SLOT']=="Free") & (VacationAirfares['GATE']=="Free") & (VacationAirfares['SW']=="Yes")]
```

```python
Find = VacationAirfares[['COUPON','FARE']]
Find['mean'] = Find.groupby('COUPON')['FARE'].transform(lambda x: x.mean())

VacationAirfares = Find.drop('FARE',axis=1).drop_duplicates().sort_values(by='mean',ascending=False).head(20)
VacationAirfares

```

```python
plt.figure(figsize=(30,30))
plt.rcParams['font.size'] = '10'

fracs = np.array(VacationAirfares['mean'])
total = sum(fracs)

colors = sns.color_palette('bright')[0:4]

plt.pie(x=VacationAirfares['mean'],
        labels=VacationAirfares['COUPON'], 
        colors  = colors,
        autopct='%.0f%%')

# add a circle at the center to transform it in a donut chart
circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)

plt.show()
```
